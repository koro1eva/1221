## Задание 4.1

```python
import pandas as pd
df = pd.read_csv('precious_metal.csv', sep=';') 
print(df.info())
```

## Задание 4.2


```python
df.set_axis(['gold', 'silver', 'platinum', 'palladium', 'date'], axis = 'columns', inplace = True) 
print(df.info())
```

## Задание 4.3

```python
df.isnull().sum()
df = df.fillna(0)
print(df.info())
```

## Задание 4.4


```python
def replace_symbol(x): 
 x = str(x) 
 x = float(x.replace(',','.')) 
 return x 
df['gold'] = df['gold'].apply(replace_symbol)
df['silver'] = df['silver'].apply(replace_symbol)
df['platinum'] = df['platinum'].apply(replace_symbol)
df['palladium'] = df['palladium'].apply(replace_symbol)
df
```

## Задание 4.5

```python
res_mea = df.mean()
res_mea

gold_mean = res_mea[0]
silver_mean = res_mea[1]
platinum_mean = res_mea[2]
palladium_mean = res_mea[3]

df['gold'] = df['gold'].apply(lambda x: gold_mean if x==0 else x)
df['silver'] = df['silver'].apply(lambda x: silver_mean if x==0 else x)
df['platinum'] = df['platinum'].apply(lambda x: platinum_mean if x==0 else x)
df['palladium'] = df['palladium'].apply(lambda x: palladium_mean if x==0 else x)
df.head()
```

## Задание 4.6

```python
df['gold_usd']=df['gold']/79
df['silver_usd']=df['silver']/79
df['platinum_usd']=df['platinum']/79
df['palladium_usd']=df['palladium']/79

df['gold_eu']=df['gold']/83
df['silver_eu']=df['silver']/83
df['platinum_eu']=df['platinum']/83
df['palladium_eu']=df['palladium']/83

df.to_csv('pr_met_new.csv')
df
```
